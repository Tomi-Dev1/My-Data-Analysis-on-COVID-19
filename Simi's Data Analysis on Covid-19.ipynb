```python
colors = ['cyan', 'green', 'orange', 'red', 'purple','blue']  # Color palette
for i, column in enumerate (['Serious,Critical', 'TotalDeaths', 'TotalRecovered', 'ActiveCases', 'TotalCases']):
    fig = px.bar(world_data.iloc[0:20], x='Country/Region', y=column, title=f'{column} by Country')
    fig.update_traces(marker_color=colors[i])   # Set specific color for each plot
    fig.show()
```

```python
#this gives a stacked bar chart of the metrics by country.
fig = go.Figure()
metrics = ['Serious,Critical', 'TotalDeaths', 'TotalRecovered', 'ActiveCases', 'TotalCases']
colors = px.colors.qualitative.Set1  # Set the color palette
for i, metric in enumerate(metrics):
    fig.add_trace(go.Bar(
        x=world_data['Country/Region'].iloc[0:20],
        y=world_data[metric].iloc[0:20],
        name=metric,
        marker_color=colors[i]  # Use the color from the palette
    ))
fig.update_layout(
    barmode='stack',
    title="Stacked COVID-19 Metrics by Country",
    xaxis_title="Country/Region",
    yaxis_title="Count"
)
fig.show()
    
```
